```python
################ Histograma ###################

import cv2
import matplotlib.pyplot as plt

#leitura da imagem
image4 = cv2.imread("image-1.jpg")
image4 = cv2.cvtColor(image4, cv2.COLOR_BGR2RGB)

#exibire o histograma de todos os pixels na imagem colorida
plt.hist(image4.ravel(), bins = 256, range = [0,256])
plt.show()
```

```python
#mostra a imagem RGB

plt.figure(figsize=(5,2))
plt.imshow(image4)
plt.title("Lago Europeu")
```

```python
#exibir o histograma do canal R, G, B.
color = ('b', 'g', 'r')
for i, col in enumerate(color):
  histr = cv2.calcHist([image4], [i], None,[256], [0,256])
  plt.plot(histr, color = col)
  plt.xlim([0,256])
  plt.show()
```

```python
#aumentar o brilho da imagem
img = cv2.imread("image-3.jpg", 0)

for i in range(100):
  for j in range(149):
    if (img[i,j] < 230):
       img[i,j]=img[i,j] + 25

cv2.imwrite("image-3.jpg", img)

image = cv2.imread("image-3.jpg")
cv2_imshow(image)
```

```python
# Criando imagens com python

'''imagem_color = np.zeros((1024, 1024, 3), np.uint8)'''
imagem_color = np.ones((1024, 1024, 3), np.uint8) * 255
plt.imshow(imagem_color)
```

```python
cv2.line(imagem_color, (0,0), (1024,1024), (0,0,255), 20)

plt.imshow(imagem_color)
plt.title("Linha Diagonal")
```

```python
cv2.line(imagem_color, (512,0), (512,1024), (255,0,0), 10)

plt.imshow(imagem_color)
plt.title("Linha Vertifcal")
```

```python
cv2.line(imagem_color, (0,512), (1024,512), (0,255,0), 30)

plt.imshow(imagem_color)
plt.title("Linha Horizontal")
```

```python
#criando um Quadrado
imagem_color = np.zeros((1024, 1024, 3), np.uint8)
cv2.rectangle(imagem_color, (100,100), (924,924), (0,255,0), 20)

plt.imshow(imagem_color)
plt.title("Quadrado")
```

```python
#criando um Quadrado preenchido
imagem_color = np.zeros((1024, 1024, 3), np.uint8)
cv2.rectangle(imagem_color, (100,100), (924,924), (255,0,0), -1)
#adicionamos valor negativo para preenchimento do desenho

plt.imshow(imagem_color)
plt.title("Quadrado Preenchido")
```

```python
#criando um Circulo
imagem_color = np.zeros((1024, 1024, 3), np.uint8)
cv2.circle(imagem_color, (512,512), 400, (255,0,0), 10)

plt.imshow(imagem_color)
plt.title("Circulo")
```

```python
#Desenhando Polígonos

imagem_color = np.zeros((512, 512, 3), np.uint8)
pts = np.array([[200,120], [300,50], [350,300], [300,400], [200,150]], np.int32)
cv2.polylines(imagem_color, [pts], True, (255, 0, 0), 10)

plt.imshow(imagem_color)
plt.title("Polígonos")
```

```python
#Desenhando Polígonos abertos

imagem_color = np.zeros((512, 512, 3), np.uint8)
pts = np.array([[200,120], [300,50], [350,300], [300,400], [200,150]], np.int32)
cv2.polylines(imagem_color, [pts], False, (255, 0, 0), 10)

plt.imshow(imagem_color)
plt.title("Polígonos")
```

```python
imagem_color = np.zeros((512, 512, 3), np.uint8)
pts = np.array([[200,120], [300,50], [350,300], [300,400], [200,150]], np.int32)
cv2.fillPoly(imagem_color, [pts], (255, 0, 0))

plt.imshow(imagem_color)
plt.title("Polígonos")
```
